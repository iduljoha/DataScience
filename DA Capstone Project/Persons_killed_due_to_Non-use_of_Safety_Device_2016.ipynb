```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Persons_killed_due_to_Non-use_of_Safety_Device_2016.csv')
```

```python
df
```

```python
df.head()
```

```python
total_deaths_male = df['Non-wearing of Helmet - Male'].sum()

total_deaths_female = df['Non-wearing of Helmet - Female'].sum()
```

```python
total_deaths_male
```

```python
total_deaths_female
```

```python
Total_Death=total_deaths_female+total_deaths_male
Total_Death
```

```python

male_percentage = (total_deaths_male / Total_Death) * 100
female_percentage = (total_deaths_male / Total_Death) * 100

print(f"Percentage of male deaths due to non-wearing of helmet: {male_percentage:.2f}%")
print(f"Percentage of female deaths due to non-wearing of helmet: {female_percentage:.2f}%")
```
