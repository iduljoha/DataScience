```python
monthly_sales = df['Sales'].resample('M').sum()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales, marker='o', linestyle='-', color='b')
plt.title('Monthly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45) 
plt.tight_layout() 
```

```python
monthly_profit=df['Profit'].resample('M').sum()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_profit.index, monthly_profit, marker='o', linestyle='-', color='r')
plt.title('Monthly Profit Trends')
plt.xlabel('Date')
plt.ylabel('Total Profit')
plt.grid(True)
plt.xticks(rotation=45) 
plt.tight_layout()  
```

```python
total_sales_by_category = df.groupby('Product Category')['Sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
total_sales_by_category.plot(kind='bar', color='b')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout() 
```

```python
total_profit_by_category = df.groupby('Product Category')['Profit'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
total_profit_by_category.plot(kind='bar', color='r')
plt.title('Total Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
```

```python
monthly_sales = df['Sales'].resample('M').sum()
```

```python
plt.figure(figsize=(14, 7))


plt.subplot(2, 1, 1)
plt.plot(monthly_sales.index, monthly_sales, marker='o', linestyle='-', color='b')
plt.title('Monthly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
```

```python
yearly_sales = df['Sales'].resample('Y').sum()
```

```python
plt.subplot(2, 1, 2)
plt.plot(yearly_sales.index, yearly_sales, marker='o', linestyle='-', color='g')
plt.title('Yearly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
```

```python
sales_by_ship_mode = df.groupby('Ship Mode')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sales_by_ship_mode.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Total Sales')
plt.xticks(rotation=30)
plt.grid(axis='y')
plt.tight_layout()
```

```python
monthly_sales_by_ship_mode = df.groupby(['Ship_Mode', pd.Grouper(freq='M')])['Sales'].sum().unstack(level=0)


plt.figure(figsize=(14, 8))
monthly_sales_by_ship_mode.plot(kind='line', marker='o')
plt.title('Monthly Sales Trends by Ship Mode')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.legend(title='Ship Mode')
plt.xticks(rotation=45)
plt.tight_layout()


```
